```python
from firebase import firebase
import time

# Initialize Firebase connection
FBconn = firebase.FirebaseApplication('https://firstapp-8067b-default-rtdb.firebaseio.com/', None)
def enterWord():
  words = input("Enter words: ")
  wordsArr = words.split()

  for word in wordsArr:
    data_to_upload = {
        'word': word
    }
    result = FBconn.post('/myTest1/', data_to_upload)

  print(result)

def get_all_records():
    """Retrieve all temperature records"""
    return FBconn.get('/myTest1/', None)

def update_temperature(record_id, new_temp):
    """Update an existing temperature record"""
    return FBconn.put('/myTest1/' + record_id, 'Temp', new_temp)

def delete_record(record_id):
    return FBconn.delete('/myTest1/', record_id)

def display_records():
    """Display all temperature records"""
    records = get_all_records()
    if records:
        print("\nCurrent Records:")
        for record_id, data in records.items():
            print(f"word : {data['word']}")
    else:
        print("\nNo records found")

while True:
    print("\nTemperature Tracker Menu:")
    print("1. Add one word")
    print("2. Add some words")
    print("3. display how many time the word is appear")
    print("4. remove word")
    print("5. display all words")

    choice = input("\nEnter your choice (1-5): ")

    if choice == '1':
      enterWord()

    elif choice == '2':
      enterWord()
    elif choice == '3':
      word = input("\nEnter the word you want to calculate how many time is appear ")
      records = get_all_records()
      count = 0
      for record_id, data in records.items():
        if data.get('word') == word:
          count+=1
      print(f"The word {word} is appear {count} times")

    elif choice == '4':
        # View all records
        deleted_word = input("\nEnter the word you want to delete ")
        result = delete_record(deleted_word)
        if result is None:
            print("Record deleted successfully!")
        else:
            print("Failed to delete record. Check if the ID exists.")


    elif choice == '5':
        display_records()


    else:
        print("Invalid choice. Please try again.")
```
